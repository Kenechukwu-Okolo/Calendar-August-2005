```python
# Python program to display calendar of August 2005
 
# import module
import calendar
 
# define variables
yr = 2005
mnth = 8
 
# select first day of the week
calendar.setfirstweekday(6)

# display the calendar
print(calendar.month(yr, mnth))
```
